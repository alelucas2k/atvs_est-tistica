1. Liste as diferenças entre uma lista, um dicionário, um array numpy e uma series pandas

Diferenças:

. Listas possuem conjuntos de valores ordenados e indexados por posição, aceitando diferentes tipos de elementos
. Dicionparios não possuem uma ordenação e o acesso aos elementos é feito por chaves 
. Arrays numpy são mais leves em termos de memória e permite fazer operaçãoes em todos os itens ao mesmo tempo
. As Series Pandas podem receber qualquer tipo de elemento e tem uma ímoca dimensão, diferente dos array numpy que são multidimensionais
